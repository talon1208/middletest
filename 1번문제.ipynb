다음 코드에서는 그룹을 이용해 치폴레의 총 판매 액수가 가장 높은 상위 10개의 항목을 출력했습니다.

```python
# -*- coding: utf-8 -*-

import pandas as pd

# read_csv 함수로 데이터를 Dataframe 형태로 불러옵니다.
file_path = '../data/chipotle.tsv'
chipo = pd.read_csv(file_path, sep = '\t')
```

```python
# order_id는 중복해서 계산되므로, 총 판매액 계산을 위해 order_id를 그룹으로 묶는다.
test = chipo.groupby('item_name').sum()
# 총 판매액이 높은 상위 10개 메뉴를 출력한다.
test= test.sort_values(['item_price'], ascending=False)
test.head(10)
```
